```python
import glob
path_xls = '/content/drive/MyDrive/Colab Notebooks/06 Lumon/Electromuebles/FilesView/'
path_result = '/content/drive/MyDrive/Colab Notebooks/06 Lumon/Electromuebles/Result/'
all_files = glob.glob(path_xls + "*.xlsx")
print(all_files)
```

```python
print(all_files[0].split('/')[-1])
```

```python
import os

now = datetime.now()
file_date = now.strftime("%Y-%m-%d_%H-%M")

folder_name = "{}/{}/".format(path_result, datetime.now().strftime('%Y-%m-%d'))
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
 

df_files = []
profiles = []
for item in all_files:
    print('Report: ', item)
    df_view = pd.read_excel(item)
    title_name = item.split('/')[-1].replace('.xlsx','').lower()
    date_file_name = "{}/ProfileReport_{}_{}.html".format(folder_name, datetime.now().strftime('%Y-%m-%d'), title_name)
    
    df_files.append(df_view)
    profile = ProfileReport(df_view, title='Profiling Report: ' + title_name, html={'style':{'full_width':True}})
    profile.to_file(output_file=date_file_name) 
    profiles.append(profile)
    profile
```

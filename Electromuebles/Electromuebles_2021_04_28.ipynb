```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
from datetime import datetime
```

```python
path_data = "/content/drive/MyDrive/Colab Notebooks/06 Lumon/Electromuebles/"
files_data = ['/content/drive/MyDrive/Colab Notebooks/06 Lumon/Electromuebles/Vista_minable.csv']

now = datetime.now()
file_date = now.strftime("%Y-%m-%d_%H-%M")

df_files = []
profiles = []
for item in files_data:
    print("{}".format(item))
    df = pd.read_csv("{}".format(item), delimiter=";", encoding='utf8')
    # pd.read_excel("{}".format(item), index_col=0)
    print(item)
    title_name = item.replace('.csv','').lower()
    df_files.append(df)
    profile = ProfileReport(df, title='Profiling Report: ' + title_name, html={'style':{'full_width':True}})
    profile.to_file(output_file='report/{}_{}.html'.format(file_date, title_name)) 
    profiles.append(profile)
```
